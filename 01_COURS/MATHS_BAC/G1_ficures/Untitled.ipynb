# 1. Les blocs et instructions de base <a name="base"></a>

Dans Scratch, on trouve quelques _blocs_ de base, pour avancer ou faire tourner le chat :

En Python, les équivalents de ces blocs de base sont les _instructions_ suivantes, qui permettent de faire avancer ou tourner une tortue (vous l’avez peut-être vue au milieu dans la figure du début) :
* `forward(x)` pour faire avancer de `x` pixels
* `right(a)` pour faire tourner de `a` degrés vers la droite
* `left(a)` pour faire tourner de `a` degrés vers la gauche

Pour exécuter plusieurs instructions à la suite, on les écrit chacune sur une ligne.<br />C’est l’équivalent d’imbriquer les blocs les uns à la suite des autres.

Par exemple :
```python
forward(100)
right(90)
forward(50)
left(45)
forward(200)
```

<div class="alert alert-warning" style="border-left:15px solid #8a6d3b;">
    En lycée, Python servira à tracer des dessins, mais aussi à faire des calculs, tracer des fonctions, choisir des nombres aléatoires, etc.<br />İl faut donc expliquer en première ligne à Python ce que l’on veut faire. <b>Pour dessiner avec la tortue, on «&nbsp;charge le <i>module</i> <span style="font-family:courier new;font-size:115%">turtle</span>&nbsp;»&nbsp;:</b>
</div>

````python
from turtle import *

forward(100)
right(90)
forward(50)
left(45)
forward(200)

done()
````

</div>

C’est joli, non ?

# 2. Répéter des instructions <a name="for"></a>

Pour dessiner un carré, il y a déjà beaucoup d’instructions, alors pour un octogone… !

Heureusement, on peut demander à Python de répéter des instructions plusieurs fois, comme avec Scratch :

<div class="alert alert-warning" style="border-left:15px solid #8a6d3b;">
    Le fonctionnement est similaire en Python. Les blocs ci-dessus donnent les instructions ci-dessous.<br />
    Remarquez que les blocs bleus contenus dans la boucle de répétition orange sont décalés.<br/>
    On retrouve ce même décalage de quelques espaces en Python.
</div>

````python
from turtle import *

for i in range(4):
    forward(50)
    right(90)

done()
````

On peut donc définir une nouvelle fonction `spirographe(k, n)` qui va tracer $k$ polygones à $n$ côtés.

On commence par la définir (bien penser à exécuter les instructions ci-dessous avec [Maj]+[Entrée]) :

````python
def spirographe(k, n):
    for fig in range(k):
        polygone(n)
        right(360 / k)
````

Rien n’est affiché : c’est normal, car il n’y a pas `done()` à la fin.

Puis il faut exécuter :

````python
spirographe(6, 5)
done()
````
