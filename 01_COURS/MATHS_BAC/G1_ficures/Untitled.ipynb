# 1. Les blocs et instructions de base <a name="base"></a>

Dans Scratch, on trouve quelques _blocs_ de base, pour avancer ou faire tourner le chat :

En Python, les √©quivalents de ces blocs de base sont les _instructions_ suivantes, qui permettent de faire avancer ou tourner une tortue (vous l‚Äôavez peut-√™tre vue au milieu dans la figure du d√©but) :
* `forward(x)` pour faire avancer de `x` pixels
* `right(a)` pour faire tourner de `a` degr√©s vers la droite
* `left(a)` pour faire tourner de `a` degr√©s vers la gauche

Pour ex√©cuter plusieurs instructions √† la suite, on les √©crit chacune sur une ligne.<br />C‚Äôest l‚Äô√©quivalent d‚Äôimbriquer les blocs les uns √† la suite des autres.

Par exemple :
```python

```
