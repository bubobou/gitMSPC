Parmi les programmes définis ci-dessous, lesquels correspondent au programme de calcul initial ?

```python
# programme 1
def programme_de_calcul1(nombre):
    resultat = 3*nombre
    resultat = resultat + 1
    resultat = resultat / 2
    return resultat
```
