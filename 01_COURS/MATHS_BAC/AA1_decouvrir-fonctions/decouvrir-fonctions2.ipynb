### Situation : programmation en python

Parmi les programmes définis ci-dessous, lesquels correspondent au programme de calcul :

 - recevoir un nombre
 - multiplier un nombre par 3
 - ajouter 1 au résultat
 - diviser le résultat par 2
 - retourner le résultat
 

```python
# programme n°1
def programme_de_calcul1(nombre):
    resultat = 3*nombre
    resultat = resultat + 1
    resultat = resultat / 2
    return resultat
```

```python
# programme n°2
def programme_de_calcul2(x):
    resultat = 3*x
    resultat = resultat + 1
    resultat = resultat / 2
    return resultat
```

```python
# programme n°3
def fonction(antecedent):
    image = 3*antecedent
    image = image + 1
    image = image / 2
    return image
```

#### Comment utiliser les programmes ?

Il faut écrire le nom du programme (le terme approprié est *fonction*) et lui donner un nombre à calculer :

*nom_de_la_fonction(**nombre_donné**)*

Par exemple ci-dessous on utilise la fonction *programme_de_calcul1* avec le nombre -2:

```python
programme_de_calcul1(-2)
```
